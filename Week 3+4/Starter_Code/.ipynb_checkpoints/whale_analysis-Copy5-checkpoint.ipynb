### 8. Calculate daily returns and store in new column

Equation: $r=\frac{{p_{t}} - {p_{t-1}}}{p_{t-1}}$

The daily return is the (current closing price minus the previous day closing price) all divided by the previous day closing price. The initial value has no daily return as there is no prior period to compare it with. 

Here the calculation uses the python shift function 

```python

df_sr['SnP_TSX_60_Returns'] = (df_sr['Close'] - df_sr['Close'].shift(1))/ df_sr['Close'].shift(1)
```

```python
df_sr.head(10)
```

### 9. Cross check conversion to daily returns against alternative method - pct_change function

```python
df_sr['SnP_TSX_60_Returns'] = df_sr['Close'].pct_change()
df_sr.head(10)
```

#### Methods cross check - looks good - continue

```python
# 
df_sr.isna().sum()
```

#### Calculate and Plot cumulative returns.

```python
def get_cumulative_return(in_series):
    '''Takes a dataframe and returns a corresponding series
    with cumulative return'''
    
    cumulative_series = in_series + in_series.shift(1) - 1
    
    return cumulative_series
    
```

```python
merged_analysis_cum_df = merged_analysis_df.copy()



for column in merged_analysis_cum_df.columns:
    #print(merged_analysis_df[column])
    print(column)
    
    # Create a new column
    col_new_str = column.replace('Returns','CumReturns') 
    print(col_new_str)
    
    # call function to geet cumulative return
    merged_analysis_cum_df[col_new_str] = get_cumulative_return(merged_analysis_cum_df[column])
    
    print(merged_analysis_cum_df.head(3))
    # drop original column
    del merged_analysis_cum_df[column]
    
    
```
